# Detect Manga Text

```python
from inference import model2annotations, load_text_detector
from utils.io_utils import imread
import PIL.Image
```

```python
model_path = r'data/comictextdetector.pt'
model = load_text_detector(model_path)

img_path = r'opm_raw.png'
img = imread(img_path)

img_array, mask, yolo_boxes = model2annotations(model, img)
```

# Extract Text Regions

```python
img = PIL.Image.fromarray(img_array)

for box in yolo_boxes:
    x1, y1, x2, y2 = box
    cropped_image = img.crop((x1, y1, x2, y2))
```
